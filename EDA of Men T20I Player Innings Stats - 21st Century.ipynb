creating a dataframe that consists only stats of player when played in India

```python
df_ind = df2.loc[df2['Ground'] == 'In India']
df_ind.head()
```

```python
# sum of runs scored by each player
sum = df_ind.groupby('Innings Player')['Innings Runs Scored Num'].sum()
```

```python
#top 5 run scorers
sum.nlargest(5)
```

V Kohli has scores most runs(1120) in India.

```python
sum.nlargest(5).plot.bar()
plt.title('Most Runs Scored In India',fontsize=20)
plt.ylabel('Runs Scored')
plt.show()
```

# HYPOTHESIS 3

An individual player(RG Sharma) has played most number of matches for a particular country(India).


Hypothesis is correct

```python
#cretaing a dataframe which consists only players stats who palyed for India.
df3 = data_batting[(data_batting['Country'] == 'India')]
df3.head()
```

```python
most_matches = df3['Innings Player'].value_counts()
#top 5 players who played most number of matches
most_matches.nlargest(5)
```

RG Sharma has played most number of matches(108) for India 

```python
most_matches.nlargest(5).plot.bar()
plt.title('Most Matches played for India',fontsize=20)
plt.ylabel('Number of Innings Played')
plt.show()
```

# HYPOTHESIS 4

G Maxwell has best strike rate in overall T20I matches.

Hypothesis is false

```python
#average of palyer batting strike rate
avg = data_batting.groupby('Innings Player')['Innings Batting Strike Rate'].mean()
```

```python
#Top 5 players who has best batting strike rate 
avg.nlargest(5)
```

```python
avg.nlargest(5).plot.barh()
plt.title('Best Batting Strike rate overall',fontsize=20)
plt.xlabel('Strike Rate')
plt.show()
```

Mohammad Asif has best batting strike rate.
